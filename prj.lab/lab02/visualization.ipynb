```python
import cv2
import matplotlib.pyplot as plt
```

```python
f, axarr = plt.subplots(nrows=1, ncols=3, figsize=(8, 8))

position = 0

with open('image_numes.txt', 'r') as file:
    for image_name in file:
        if image_name == "":
            break
        image = cv2.imread(f'../../data/walnut01/{image_name.strip()}', -1)
        axarr[position].imshow(image)
        axarr[position].set_title(image_name)
        horizontal_size, vertical_size = image.shape
        axarr[position].annotate(f'size: {image.shape},\nMaximal value: {str(image.max())[:10]}', xy=(0, 0))
        position += 1
```
